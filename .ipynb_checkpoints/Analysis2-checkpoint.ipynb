```python
import os
import csv
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import stats
```

```python
input_educated_df = pd.read_csv('Data/US Education 2014.csv')
input_mortality_state_df = pd.read_csv('Data/CleanedState.csv')
input_mortality_df = input_mortality_state_df.loc[input_mortality_state_df['Year']==2014]
merged_df = pd.merge(input_educated_df, input_mortality_df, on = 'State')
merged_df.head()
```

```python
merged_scatter_df = merged_df[["Education","Mortality"]]
```

```python
mortality = merged_df['Mortality']
education = merged_df ['Education']
```

```python
plt.scatter(education, mortality, marker = 'o', facecolor = 'r', alpha=0.8, s=80, edgecolor ='black')
plt.ylabel('Mortality Rates | Deaths/ 100K')
plt.xlabel('Education Level | Percent of Graduates')
plt.title('Education Level v Mortality Rates')
plt.tight_layout()
plt.savefig('Data/EducationvMortality.png', dpi = 800)
```

```python
(slope, intercept, r_value, p_value, std_err) = stats.linregress(education, mortality)
plt.scatter(education,mortality,facecolor="r",edgecolor="black",label="Original Data",alpha=.8,s=80)
plt.plot(education,intercept+slope*education,color="blue",label="Regression Line")
plt.xlabel("Education Level | Percent of Graduates")
plt.ylabel("Mortality Rates | Deaths per 100k")
plt.title("Education Level v Mortality Rates Regression Line")
plt.legend()
print("The P-Value is " + str(p_value))
print(str(r_value*r_value))
plt.savefig("Data/EducationvsMortalityRegression.png", dpi=800)
```
