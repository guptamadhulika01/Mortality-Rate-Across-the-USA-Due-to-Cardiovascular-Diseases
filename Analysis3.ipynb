```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import stats
import numpy as np
```

```python
California = pd.read_csv('CaliforniaMortality.csv')
Mississippi=pd.read_csv('Mississippi_Annual.csv')
Minnesota=pd.read_csv('Minnesota_Annual.csv')
CaYear = California["Year"]
MinYear=Minnesota["Year"]
MisYear=Mississippi["Year"]
CaMortality = California["California Mortality"]
MinMortality=Minnesota["Mortality"]
MisMortality=Mississippi["Mortality"]
```

```python
(slope, intercept, r_value, p_value, std_err) = stats.linregress(CaYear, CaMortality)
plt.scatter(CaYear,CaMortality,facecolor="r",edgecolor="black",label="Original Data",alpha=.8,s=80)
plt.plot(CaYear,intercept+slope*CaYear,color="blue",label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Mortality Rate | Deaths per 100k")
plt.title("California Mortality Rates From 1980-2014 Regression Line")
plt.legend()
print("The P-Value is " + str(p_value))
print("y = "+ str(slope.round(2))+"(x)"  + "+" + str(intercept.round(2)))
plt.savefig("CaliforniaMortalityRegression.png", dpi=800)
```

```python
(slope, intercept, r_value, p_value, std_err) = stats.linregress(MinYear, MinMortality)
plt.scatter(MinYear,MinMortality,facecolor="r",edgecolor="black",label="Original Data",alpha=.8,s=80)
plt.plot(MinYear,intercept+slope*MinYear,color="blue",label="Regression Line")
plt.xlabel("Year")
plt.ylabel("Mortality Rate | Deaths per 100k")
plt.title("Minnesota Mortality Rates From 1980-2014 Regression Line")
plt.legend()
print("The P-Value is " + str(p_value))
print("y = "+ str(slope.round(2))+"(x)"  + "+" + str(intercept.round(2)))
plt.savefig("MinnesotaMortalityRegression.png", dpi=800)
```
